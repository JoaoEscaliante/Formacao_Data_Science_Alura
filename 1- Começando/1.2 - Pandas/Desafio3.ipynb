#1 VERIFICAR NA BASE SE CONTÉM DADOS NULOS, SE SIM, REALIZAR O TRATAMENTO

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url)
dados
```

```python
dados.isnull().sum()
```

```python
dados = dados.fillna(0)
dados
```

#2 ALICE E CARLOS NÃO FAZEM PARTE DA TURMA, LOGO, REMOVA-OS DA BASE DE DADOS

```python
remover_alunos = dados.query('Nome == "Alice" | Nome == "Carlos"').index
```

```python
# removendo as linhas referentes aos alunos "Alice" e "Carlos"
dados.drop(remover_alunos, axis=0, inplace=True)
```

```python
dados
```

```python
dados.sort_values('Nome')
```

#3 APLICAR UM FILTRO QUE SELECIONE APENAS OS ALUNOS APROVADOS

```python
#é criar uma variável onde nós realizamos a seleção apenas dos alunos aprovados:
selecao = dados['Aprovado'] == True
```

```python
dados[selecao]
```

#4 SALVAR O DF QUE POSSUI APENAS OS APROVADOS EM UM ARQUIVO CSV "ALUNOS_APROVADOS.CSV"

```python
alunos_aprovados = dados[selecao]
```

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
dados
```

#EXTRA: CONFERINDO OS ALUNOS APROVADOS, HOUVE ALGUMAS NOVAS INCORRETAS. OS QUE TEM 7,0 TÊM UM PONTO EXTRA PORÉM NÃO FOI CONTABILIZADO. COM ISSO SUBSTITUA AS NOTAS 7.0 PARA 8.0

```python
#O método replace é utilizado para substituir um determinado valor por outro.
#No caso da nossa base de dados, nós queremos substituir os valores 7.0 por 8.0. Para isso, nós podemos utilizar o seguinte trecho de código:
alunos_aprovados.replace(7.0, 8.0)
```

```python
alunos_aprovados = alunos_aprovados.replace(7.0, 8.0)
alunos_aprovados
```
